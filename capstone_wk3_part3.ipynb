<h2>Segmenting and Clustering Neighborhoods in Toronto - Part 3/3<h2>

```python
import pandas as pd
```

<h5>Obtain the data in the table of postal codes and  transform the data into a pandas dataframe<h5> 

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
df_list = pd.read_html(url, header = 0)
df_canada=df_list[0]
```

```python
df_canada=df_canada[df_canada['Borough']!='Not assigned']
```

<h5>To get the geographical coordinates of canadian postal codes using csv @
http://cocl.us/Geospatial_data<h5>

```python
df_geo=pd.read_csv('http://cocl.us/Geospatial_data')
```

<h5>Merge dataframe with coordinates

```python
df_canada = df_canada.join(df_geo.set_index('Postal Code'), on='Postal Code')
df_canada.reset_index(inplace = True) 
df_canada=df_canada.drop(['index'],axis=1)
df_canada
```

```python
df_canada.shape
```

<h5>Explore and cluster the neighborhoods in Toronto - work with Boroughs that contain the word Toronto</h5>

```python
df_toronto=df_canada[df_canada['Borough'].str.contains('Toronto')]
```

```python
import folium
from geopy.geocoders import Nominatim
```

```python
address = 'Toronto, Canada'

geolocator = Nominatim(user_agent="ca_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of New York using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_toronto['Latitude'], df_toronto['Longitude'], df_toronto['Borough'], df_toronto['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
