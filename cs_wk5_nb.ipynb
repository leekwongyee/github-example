<h5>Function to repeat the same process to all the neighborhoods in Singapore

```python
import requests

def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

<h5>Code to run the above function on each neighborhood and create a new dataframe called *sg_venues*

<h5>Analyze Each Neighborhood

```python
# one hot encoding
sg_onehot= pd.get_dummies(sg_venues[['Venue Category']], prefix="", prefix_sep="")

# drop catergory Neighborhood if it exists
if("Neighborhood" in sg_venues['Venue Category'].unique()):
    sg_onehot=sg_onehot.drop(['Neighborhood'],axis=1)

# add Location column back to dataframe
sg_onehot['Neighborhood'] = sg_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [sg_onehot.columns[-1]] + list(sg_onehot.columns[:-1])

sg_onehot = sg_onehot[fixed_columns]

sg_onehot.head()
```

<h5>Group rows by neighborhood and by taking the mean of the frequency of occurrence of each category

```python
sg_grouped = sg_onehot.groupby('Neighborhood').mean().reset_index()
sg_grouped
```

<h5>Create a new DataFrame for categories that have an impact on Food Court.
Competitor: Coffee Shop,
Ideal location for Food Court: Shopping Mall, Metro Station

```python
sg_food_court = sg_grouped[['Neighborhood','Food Court','Coffee Shop','Shopping Mall','Metro Station']]
```

```python
sg_food_court.head()
```

```python
sg_food_court_clustering = sg_food_court.drop('Neighborhood', 1)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

distortions = []
I = range(1,10)
for i in I:
    kmeans = KMeans(n_clusters=i, random_state=0).fit(sg_food_court_clustering)
    distortions.append(kmeans.inertia_)

plt.figure(figsize=(16,8))
plt.plot(I, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
    
```

<h5>Set optimal number of clusters (=5 from Elbow Method) 

```python
import numpy as np

kclusters = 5

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(sg_food_court_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
# add clustering labels

#neighborhoods_venues_sorted=neighborhoods_venues_sorted.drop(neighborhoods_venues_sorted['Cluster Labels'].axis=1)
sg_food_court.insert(0, 'Cluster Labels', kmeans.labels_)

sg_merged = loc_data

# merge neighborhoods_venues_sorted with loc_data to add latitude/longitude for each neighborhood
sg_merged = sg_merged.join(sg_food_court.set_index('Neighborhood'), on='Neighborhood')

sg_merged
```

<h5>Let's visualize the resulting clusters

```python
import folium
import matplotlib.cm as cm
import matplotlib.colors as colors


# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(sg_merged['Latitude'], sg_merged['Longitude'], sg_merged['Neighborhood'], sg_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
     
        
map_clusters
```

<h5>Create a scoring table for the 5 clusters based on the mean score of each attributes
and negate the values of competing attributes
    
#Cluster score = (‘Food Court’ + ‘Coffee Shop’)*(-1) + ( ‘Shopping Mall’+ ‘Metro Station’) 

```python
col_names =  ['Cluster', 'Food Court(-)', 'Coffee Shop(-)', 'Shopping Mall(+)', 'Metro Station(+)']
cluster_score  = pd.DataFrame(columns = col_names)

for c in np.arange(kclusters):
    cluster=sg_merged.loc[sg_merged['Cluster Labels'] == c].describe().loc[['mean']]
    clu=cluster.iloc[0,2]
    fc=cluster.iloc[0,3]*-1
    cs=cluster.iloc[0,4]*-1
    sm=cluster.iloc[0,5]
    ms=cluster.iloc[0,6]
    new_row = {'Cluster':"c"+str(int(clu)),'Food Court(-)':fc,'Coffee Shop(-)':cs,'Shopping Mall(+)':sm,'Metro Station(+)':ms}
    cluster_score = cluster_score.append(new_row, ignore_index=True)

cluster_score["Score"] = cluster_score.sum(axis=1)
cluster_score

```

```python
list=['1st','2nd','3rd','4th','5th']
cluster_rank=cluster_score.sort_values(by=['Score'],ascending=False)
cluster_rank['Rank']=list
cluster_rank.set_index('Rank', inplace = True)
cluster_rank

# Cluster score = (‘Food Court’ + ‘Coffee Shop’)*(-1) + ( ‘Shopping Mall’+ ‘Metro Station’) 
```

From the clustering exercise and based on the table above, cluster 0 is ranked the highest based on the scoring. Although having less competition, the mean value of the shopping mall indicates that the neighbourhood has a comparatively low number of shopping malls compared to other clusters.
